## Statistical Trend Analysis using  Original Mann-Kendall test (original_test): 

#### Original Mann-Kendall test is a nonparametric test, which does not consider serial correlation or seasonal effects.

```python
DTR = temporal_data.DRT_a
mk.original_test(DTR, alpha=0.05 )

### From this result, we can say that there is a significant trend in this temporal dataset. Because the p-value is smaller than alpha=0.05 and h= false. The trend is increasing and the value of trend/slope is 0.00796666666666675
```

## Using pyMannKendall package functions (Hamed and Rao Modified MK Test (hamed_rao_modification_test)  for creating a temporal figure 

#### Hamed and Rao Modified MK Test (hamed_rao_modification_test): This modified MK test proposed by Hamed and Rao (1998) to address serial autocorrelation issues. They suggested a variance correction approach to improve trend analysis. User can consider first n significant lag by insert lag number in this function. By default, it considered all significant lags. 

https://github.com/mmhs013/pyMannKendall/blob/master/Examples/Example_pyMannKendall.ipynb

```python

DTR = temporal_data.DRT_a

fig, ax = plt.subplots(figsize=(5, 4))
res = mk.hamed_rao_modification_test(DTR)
trend_line = np.arange(len(DTR)) * res.slope + res.intercept

ax.plot(DTR, color='red', marker='o')
ax.plot(DTR.index, trend_line, color='blue')
ax.legend(['DTR', 'trend line'])
plt.savefig('DTR.png', dpi = 300)


# SU = temporal_data.SU_a

# fig, ax = plt.subplots(figsize=(5, 4))
# res = mk.hamed_rao_modification_test(SU)
# trend_line = np.arange(len(SU)) * res.slope + res.intercept

# ax.plot(SU,color='red', marker='o')
# ax.plot(SU.index, trend_line,color='blue')
# ax.legend(['SU','trend line'])

# plt.savefig('SU.png', dpi = 300)


# TNm = temporal_data.TNm_a
# fig, ax = plt.subplots(figsize=(5, 4))
# res = mk.hamed_rao_modification_test(TNm)
# trend_line = np.arange(len(TNm)) * res.slope + res.intercept

# ax.plot(TNm ,color='red', marker='o')
# ax.plot(TNm.index, trend_line, color='blue')
# ax.legend(['TNm', 'trend line'])

# plt.savefig('TNm.png', dpi = 300)

# TMm = temporal_data.TMm_a
# fig, ax = plt.subplots(figsize=(5, 4))
# res = mk.hamed_rao_modification_test(TMm)
# trend_line = np.arange(len(TMm)) * res.slope + res.intercept

# ax.plot(TMm,color='red', marker='o')
# ax.plot(TXge30 .index, trend_line, color='blue')
# ax.legend(['TMm', 'trend line'])

# plt.savefig('TMm.png', dpi = 300)


# TXm = temporal_data.TXm_a
# fig, ax = plt.subplots(figsize=(5, 4))
# res = mk.hamed_rao_modification_test(TXm)
# trend_line = np.arange(len(TXm)) * res.slope + res.intercept

# ax.plot(TXm,color='red', marker='o')
# ax.plot(TXm.index, trend_line, color='blue')
# ax.legend(['TXm', 'trend line'])

# plt.savefig('TXm.png', dpi = 300)


# TXn = temporal_data.TXn_a
# fig, ax = plt.subplots(figsize=(5, 4))
# res = mk.hamed_rao_modification_test(TXn)
# trend_line = np.arange(len(TXn)) * res.slope + res.intercept

# ax.plot(TXn,color='red', marker='o')
# ax.plot(TXn.index, trend_line, color='blue')
# ax.legend(['TXn', 'trend line'])

# plt.savefig('TXn.png', dpi = 300)




```
